```python
import pandas as pd

with open('앙상블2023상승검증_코스닥.txt', 'r') as f:
    data = f.read()

# 문자열을 줄 단위로 나누고, 각 줄을 종목명과 날짜로 분리합니다.
rows = [(line[:-10].strip(), line[-10:]) for line in data.split('\n')]

# 데이터 프레임을 생성합니다.
df = pd.DataFrame(rows, columns=['종목명', '기준일자'])

duplicated_rows = df[df.duplicated()]

if not duplicated_rows.empty:
    print("중복된 행:")
    print(duplicated_rows)

# 중복된 행 제거
df = df.drop_duplicates()

# 데이터 프레임을 엑셀 파일로 저장합니다.
df.to_excel('stock_data.xlsx', index=False)
df
```
