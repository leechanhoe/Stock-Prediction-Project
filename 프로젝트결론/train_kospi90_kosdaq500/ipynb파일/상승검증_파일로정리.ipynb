```python
stock_codes = dict(kospi + kosdaq)
```

```python
df = pd.read_excel('앙상블2023상승검증.xlsx')
df
```

```python
def get_stock_data(row, code_dict=kosdaq):
    stock_code = stock_codes.get(row['종목명'])
    if stock_code is None:
        return None, None, None, None

    stock_data = fdr.DataReader(stock_code, row['기준일자'])

    if stock_data.empty:
        return None, None, None, None

    close_price = stock_data.iloc[0]["Close"]
    change = stock_data.iloc[0]["Change"]

    after_5_days_price = 0
    avg_5_days = 0
    if(5 < len(stock_data)):
        after_5_days_price = stock_data.iloc[5]["Close"]
        avg_5_days = stock_data.iloc[1:6]["Close"].mean()

    else:
        print(f"{row['종목명'], row['기준일자']} 에 향후 5일치 데이터가 없음")

    return close_price, after_5_days_price, avg_5_days, change

# 데이터프레임에 열 추가
df[['기준일종가', '5거래일후종가', '5일평균가', '변동률']] = df.apply(get_stock_data,
                                                    axis=1,
                                                    result_type="expand")
df
```

```python
df = df[(df['기준일종가'] != 0) & (df['5거래일후종가'] != 0) & (df['5일평균가'] != 0)]

if df['기준일자'].dtype == 'object':
    df['기준일자'] = pd.to_datetime(df['기준일자'])
df = df[df['기준일자'].dt.year >= 2023]

df
```

```python
def check_5_days_after_answer(row):
    if row['기준일종가'] is None or row['5거래일후종가'] is None:
        return None
    return 'O' if row['5거래일후종가'] >= row['기준일종가'] else 'X'

df['5거래일후정답여부'] = df.apply(check_5_days_after_answer, axis=1)
print(df)
```

```python
def check_5_days_mean_answer(row):
    if row['기준일종가'] is None or row['5일평균가'] is None:
        return None
    return 'O' if pd.notnull(row['5일평균가']) and row['5일평균가'] >= row['기준일종가'] else 'X'

df['5일평균정답여부'] = df.apply(check_5_days_mean_answer, axis=1)
df
```

```python
def calculate_return(row):
    if row['기준일종가'] is None or row['5거래일후종가'] is None:
        return None
    if row['5거래일후종가'] == 0:
        return 0
    return (row['5거래일후종가'] - row['기준일종가']) / row['기준일종가']

df['수익률'] = df.apply(calculate_return, axis=1)
df
```

```python
count_df = df.groupby('기준일자').size()
average_count = count_df.mean()

print(f"기준일자별 데이터 평균 개수: {average_count}")

# 각 기준일자별로 데이터의 개수 계산
count_df = df.groupby('기준일자').size()
print(count_df.unique())

# 데이터 개수가 5개 이상인 기준일자만 선택
valid_dates = count_df[count_df >= average_count].index
print(valid_dates)

# 선택한 기준일자를 가진 행만 남김
df2 = df[df['기준일자'].isin(valid_dates)]
```
